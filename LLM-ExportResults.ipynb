```python
import pandas as pd

# Assuming your test dataset is in a CSV file
try:
    test_data = pd.read_csv('test.csv', encoding='ISO-8859-1')
except UnicodeDecodeError:
    # If 'ISO-8859-1' encoding doesn't work, try 'utf-16' encoding
    test_data = pd.read_csv('test.csv', encoding='utf-16')
# Assuming your test dataset has a 'text' column containing the text data
texts = test_data['text'].tolist()
```

```python
from transformers import AutoTokenizer, AutoModelForSequenceClassification
from scipy.special import softmax

MODEL_NAME = "cardiffnlp/twitter-roberta-base-sentiment"
tokenizer = AutoTokenizer.from_pretrained(MODEL_NAME)
model = AutoModelForSequenceClassification.from_pretrained(MODEL_NAME)

sentiment_labels = ['negative', 'neutral', 'positive']

for text in texts:
    encoded_text = tokenizer(text, return_tensors='pt')
    output = model(**encoded_text)
    scores = output.logits[0].detach().numpy()
    scores = softmax(scores)

    max_score_index = scores.argmax()
    predicted_sentiment = sentiment_labels[max_score_index]

    scores_dict = {
        'negative': scores[0],
        'neutral': scores[1],
        'positive': scores[2]
    }

    print("Text:", text)
    print("Sentiment Scores:", scores_dict)
    
    if predicted_sentiment == 'negative':
        print("Predicted Sentiment: Negative")
    elif predicted_sentiment == 'neutral':
        print("Predicted Sentiment: Neutral")
    else:
        print("Predicted Sentiment: Positive")
    print("\n")
```

```python
import pandas as pd
from transformers import AutoTokenizer, AutoModelForSequenceClassification
from scipy.special import softmax

MODEL_NAME = "cardiffnlp/twitter-roberta-base-sentiment"
tokenizer = AutoTokenizer.from_pretrained(MODEL_NAME)
model = AutoModelForSequenceClassification.from_pretrained(MODEL_NAME)

sentiment_labels = ['negative', 'neutral', 'positive']

# Read the test dataset from 'test.csv'
try:
    test_data = pd.read_csv('test.csv', encoding='ISO-8859-1')  # Try different encodings if needed
except UnicodeDecodeError:
    test_data = pd.read_csv('test.csv', encoding='utf-16')

texts = test_data['text'].tolist()

# Create empty lists to store the results
output_texts = []
scores = []
sentiments = []

for text in texts:
    # Check if text is a string
    if isinstance(text, str):
        encoded_text = tokenizer(text, return_tensors='pt')
        output = model(**encoded_text)
        scores_arr = output.logits[0].detach().numpy()
        scores_arr = softmax(scores_arr)

        max_score_index = scores_arr.argmax()
        predicted_sentiment = sentiment_labels[max_score_index]

        output_texts.append(text)
        scores.append(scores_arr[max_score_index])
        sentiments.append(predicted_sentiment)
    else:
        # Handle non-string values if needed
        print(f"Skipping non-string value: {text}")

# Create a DataFrame to store the results
data = {
    'Text': output_texts,
    'Sentiment_Score': scores,
    'Sentiment_Label': sentiments
}
output_df = pd.DataFrame(data)

# Save the DataFrame to a CSV file
output_df.to_csv('sentiment_results.csv', index=False)
```
